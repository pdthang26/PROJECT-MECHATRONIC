```python
from ultralytics import YOLO
import cv2
import numpy as np
import torch
```

```python
model_seg = YOLO('YOLO_pretrain_models\yolov8n-seg.pt')


predict = model_seg("YOLO_pretrain_models\\test.png",stream=True,device = '0')
for r in predict:
    masks  = r.masks
# masks[0].cpu().shape
result_array = (masks.cpu().data.numpy() *255).astype("uint8")
        # image processing
H, W = result_array.shape[1],result_array.shape[2]
print(H,W)

```

```python
# Load the YOLOv8 model
model = YOLO('yolov8n-seg.pt')

# Open the video file
video_path = "video.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame, device = '0')

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python

combined_image = np.zeros((height , width), dtype=np.uint8)
for i in range(len(predict[0].masks)):
    combined_image += (predict[0].masks[i].data.numpy() *255).astype("uint8").squeeze(axis=0) 
    combined_image[combined_image>=255]=255
frame_resize = cv2.warpPerspective(combined_image,M_resize,(width_resize,height_resize))
print(frame_resize.shape)
cv2.imshow('Result Image', frame_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
